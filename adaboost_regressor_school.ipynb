```python
import pandas as pd
pd.set_option('display.max_columns', 33)
escola = pd.read_csv('/content/school_grades_dataset.csv')
```

```python
escola.head()
```

```python
escola.shape
```

```python
faltantes = escola.isnull().sum()
faltantes_percentual = (faltantes/len(escola['sex']))*100
print(faltantes_percentual)
```

```python
escola.dtypes
```

```python
objetos = escola.drop(['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc',
                      'Walc', 'health', 'absences', 'G1', 'G2', 'G3'], axis = 1)
```

```python
objetos_encoding = pd.get_dummies(objetos)
```

```python
objetos_encoding.head()
```

```python
escola.drop(['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup',
            'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic'], axis = 1, inplace = True)
```

```python
concatenado = pd.concat([escola, objetos_encoding], axis = 1)
```

```python
concatenado.shape
```

```python
x = concatenado.drop('G3', axis = 1)
y = concatenado['G3']
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

modelo = AdaBoostRegressor()
kfold = KFold(n_splits = 5)
resultado = cross_val_score(modelo,x,y,cv = kfold, n_jobs = -1)
print(resultado.mean())
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

modelo = AdaBoostRegressor(n_estimators = 500, learning_rate = 0.1)
kfold = KFold(n_splits = 5)
resultado = cross_val_score(modelo,x,y,cv = kfold, n_jobs = -1)
print(resultado.mean())
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

valores_grid = {'learning_rate': np.array([0.3,0.2,0.1,0.05])}

modelo1 = AdaBoostRegressor(n_estimators = 500)

gridAdaBoost = GridSearchCV(estimator = modelo1, param_grid = valores_grid, cv = 3, n_jobs = -1)
gridAdaBoost.fit(x,y)

print('Melhor taxa de aprendizagem: ', gridAdaBoost.best_estimator_.learning_rate)
print('Acurácia: ', gridAdaBoost.best_score_)
```
